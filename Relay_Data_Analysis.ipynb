## Claimed Before?

```python
user_data['Claimed Before'].value_counts()
```

```python
user_data['Claimed Before'].value_counts().index.tolist()
```

```python
import plotly.graph_objects as go

values = list(user_data['Claimed Before'].value_counts())
labels_clean = user_data['Claimed Before'].value_counts().index.tolist()
fig = go.Figure(data=[go.Pie(values=values, 
                            labels=labels_clean,
                            texttemplate = "%{label}: %{value:} users<br>(%{percent})")])

fig.update_layout(title_text="Has the user claimed before?")
```
